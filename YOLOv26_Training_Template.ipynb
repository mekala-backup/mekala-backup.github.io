# YOLOv8 Segmentation Training (Resume Checkpoint)

This notebook resumes training from a previous `last.pt` checkpoint.

## Before running
1. Install dependencies.
2. Update the two paths below:
   - `resume_weights_path`
   - `data_yaml_path`
3. Run all cells.

```
import os
from ultralytics import YOLO

# ============================
# USER CONFIGURATION
# ============================

# Path to the last checkpoint
resume_weights_path = "/path/to/runs/segment/experiment/weights/last.pt"

# Path to your dataset YAML
data_yaml_path = "/path/to/data.yaml"

# Training parameters
IMG_SIZE = 640
EPOCHS = 150
BATCH = -1          # Auto batch size
DEVICE = 0          # GPU index (use 'cpu' for CPU)
WORKERS = 4

PROJECT = "yolo_segment_training"
RUN_NAME = "resume_training"

# Verify checkpoint exists
if not os.path.exists(resume_weights_path):
    raise FileNotFoundError(
        f"Checkpoint not found: {resume_weights_path}"
    )

print("Loading model...")
model = YOLO(resume_weights_path)
```

```
results = model.train(
    data=data_yaml_path,
    imgsz=IMG_SIZE,
    epochs=EPOCHS,
    batch=BATCH,
    device=DEVICE,
    workers=WORKERS,
    cache=False,
    optimizer="AdamW",
    lr0=0.001,
    cos_lr=True,
    mosaic=0.7,
    mixup=0.1,
    copy_paste=0.2,
    close_mosaic=15,
    project=PROJECT,
    name=RUN_NAME,
    exist_ok=True,
    resume=True,
)

print("Training completed!")
print("Results saved to:", results.save_dir)
```
